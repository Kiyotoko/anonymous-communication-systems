## How it works

A VPN hides a IP address by letting the network redirect it through a specially configured remote server run by a VPN host.

## Weaknesses

- Lack of accountability
- False sense of security

````python
"""
It is assumed that when a VPN server is taken over, all 
users are also de-anonymised. In addition, it is assumed 
that the users' data traffic is distributed evenly across 
all servers and that they no longer change their server 
after a connection has been established.
"""

import matplotlib.pyplot as plt
import numpy

x = numpy.linspace(0, 1, 100)
f = lambda x: x

plt.xlabel("Corrupted Servers [%]")
plt.ylabel("Deanonymised Users [%]")
plt.plot(x, f(x))
````
